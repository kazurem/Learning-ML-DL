```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
plt.style.use("ggplot")
```

```python
df = pd.read_csv("./data/data.csv")
```

```python
plt.scatter(df['x'], df['y'], s=6)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs x")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    np.array(df['x']), np.array(df['y']), test_size=0.33, random_state=42)
```

```python
linear_model = tf.keras.Sequential(
    [
        tf.keras.Input(shape=(1, 1), name="input"),
        tf.keras.layers.Dense(units=1, activation="linear", name="output"),
    ]
)

linear_model.compile(
    loss=tf.keras.losses.MeanSquaredError(),
)
```

```python
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)

X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
history = linear_model.fit(X_train, y_train, epochs=100)
```

```python
plt.plot(np.arange(100), history.history['loss'])
plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("loss over epochs")
```

```python
pred = linear_model.predict(X_test)
```

```python
pred = pred.reshape(-1, )
```

```python
plt.scatter(df['x'], df['y'], s=6)
plt.plot(X_test, pred, c="black")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs x")
```
